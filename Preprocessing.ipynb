Ler arquivo csv e manter apenas as colunas necessárias

```python
import pandas as pd

df_23 = pd.read_csv("https://raw.githubusercontent.com/arturrur/mc853/refs/heads/main/source/csv/LEPTBR23.csv", usecols=["SEM_NOT", "SG_UF_NOT", "DT_SIN_PRI", "NU_IDADE_N", "CS_RACA",
                                                                                                                      "CS_SEXO", "CS_ESCOL_N", "CS_GESTANT", "NDUPLIC_N", "ANT_CB_LAM", "ANT_CB_CRI",
                                                                                                                      "ANT_CB_CAI", "ANT_CB_FOS", "ANT_CB_SIN", "ANT_CB_PLA", "ANT_CB_COR",
                                                                                                                      "ANT_CB_ROE", "ANT_CB_GRA", "ANT_CB_TER", "ANT_CB_LIX", "ANT_HUMANO", "ANT_ANIMAI",
                                                                                                                      'CLI_FEBRE', 'CLI_MIALGI', 'CLI_CEFALE', 'CLI_PROST', 'CLI_CONGES', 'CLI_PANTUR',
                                                                                                                      'CLI_VOMITO', 'CLI_DIARRE', 'CLI_ICTERI','CLI_RENAL', 'CLI_RESPIR',
                                                                                                                      'CLI_CARDIA', 'CLI_HEMOPU', 'CLI_HEMORR', 'CLI_MENING',
                                                                                                                      "CLI_DT_ATE", "EVOLUCAO"])

df_22 = pd.read_csv("https://raw.githubusercontent.com/arturrur/mc853/refs/heads/main/source/csv/LEPTBR22.csv", usecols=["SEM_NOT", "SG_UF_NOT", "DT_SIN_PRI", "NU_IDADE_N", "CS_RACA",
                                                                                                                      "CS_SEXO", "CS_ESCOL_N", "CS_GESTANT", "NDUPLIC_N", "ANT_CB_LAM", "ANT_CB_CRI",
                                                                                                                      "ANT_CB_CAI", "ANT_CB_FOS", "ANT_CB_SIN", "ANT_CB_PLA", "ANT_CB_COR",
                                                                                                                      "ANT_CB_ROE", "ANT_CB_GRA", "ANT_CB_TER", "ANT_CB_LIX", "ANT_HUMANO", "ANT_ANIMAI",
                                                                                                                      'CLI_FEBRE', 'CLI_MIALGI', 'CLI_CEFALE', 'CLI_PROST', 'CLI_CONGES', 'CLI_PANTUR',
                                                                                                                      'CLI_VOMITO', 'CLI_DIARRE', 'CLI_ICTERI','CLI_RENAL', 'CLI_RESPIR',
                                                                                                                      'CLI_CARDIA', 'CLI_HEMOPU', 'CLI_HEMORR', 'CLI_MENING',
                                                                                                                      "CLI_DT_ATE", "EVOLUCAO"])

df_23["DT_SIN_PRI"] = pd.to_datetime(df_23["DT_SIN_PRI"])
df_23["CLI_DT_ATE"] = pd.to_datetime(df_23["CLI_DT_ATE"])
df_22["DT_SIN_PRI"] = pd.to_datetime(df_22["DT_SIN_PRI"])
df_22["CLI_DT_ATE"] = pd.to_datetime(df_22["CLI_DT_ATE"])
```

Juntar df de 2022 com df de 2023

```python
df = pd.concat([df_22, df_23], ignore_index=True)
df
```

EVOLUCAO:
- Remover linhas do df que não tenham valor
- Remover linhas que tenham valor ignorado

```python
df = df.dropna(subset=["EVOLUCAO"])
df = df[df["EVOLUCAO"] != 9.0]
df
```

NDUPLIC_N:
- Remover duplicatas
- Remover a coluna NDUPLIC_N, que não tem mais uso

```python
df = df[df["NDUPLIC_N"] != 2.0]
df = df.drop(columns=["NDUPLIC_N"])
df
```

SG_UF_NOT:
- Criar bitmap a partir da coluna SG_UF_NOT
- Deletar coluna SG_UF_NOT

```python
regions ={
    1: "Norte",
    2: "Nordeste",
    3: "Sudeste",
    4: "Sul",
    5: "Centro-Oeste"
}
df["int"] = df["SG_UF_NOT"] // 10
df["region"] = df["int"].map(regions)
df = pd.get_dummies(df, columns=['region'], prefix='', prefix_sep='')
df = df.drop(columns=["int", "SG_UF_NOT"])

print(df)
```

SEM_NOT:
- Reformatar SEM_NOT, que está na forma aaaass, para apenas ss (semana).
- Como estamos com dados de 2022 e 2023, vamos fazer SEM_NOT ir de -1 até 105. Ou seja, última semana de 2021 até primeira semana de 2024

Obs: Teve uma amostra que tinha o valor de SEM_NOT como 2350, muito provavelmente um erro de digitação para 202350, ajustamos seu valor

```python
df.loc[df['SEM_NOT'] == 2350, 'SEM_NOT'] = 202350
df["SEM_NOT"] = df["SEM_NOT"].apply(lambda x: (x % 100) + (-53 if x // 100 == 2021 else (52 if x // 100 == 2023 else (104 if x // 100 == 2024 else 0))))
df
```

CLI_DT_ATE & DT_SIN_PRI:
- Criar uma coluna que representa a quantidade de dias entre primeiros sintomas e data de atendimento.
- Tirar dos dados aqueles que têm mais de 20 dias entre as duas datas.
- Remover as colunas CLI_DT_ATE e DT_SIN_PRI, que não têm mais uso.

```python
df['DIF_SINTOMAS_ATEND'] = (df['CLI_DT_ATE'] - df['DT_SIN_PRI']).dt.days
df = df[(df['DIF_SINTOMAS_ATEND'] <= 20) & (df['DIF_SINTOMAS_ATEND'] >= 0)]
df = df.drop(columns=["CLI_DT_ATE", "DT_SIN_PRI"])
df
```

CS_GESTANT, transformar a coluna em bool
- True se gestante
- False caso contrário

```python
df['CS_GESTANT'] = df['CS_GESTANT'].apply(lambda x: True if x in [1, 2, 3, 4] else False)
df
```

NU_IDADE_N:
- Remover todos que têm menos que 7 ano de idade, pois não existe um valor de escolaridade para eles.
- Reformatar a coluna de idade para mostrar apenas quantos anos eles têm

```python
df = df[df['NU_IDADE_N'] // 1000 == 4]
df.loc[:, 'NU_IDADE_N'] = df['NU_IDADE_N'] % 1000
df = df[df['NU_IDADE_N'] >= 7]
df
```

CS_RACA:
- Remover dados faltantes
- Remover raças com poucos dados (amarela e indígena)
- Criar um bitmap pras raças

```python
racas = {
    1: "Branca",
    2: "Preta",
    3: "Amarela",
    4: "Parda",
    5: "Indigena",
    9: "Ignorado"
}

df = df.dropna(subset=["CS_RACA"])
df = df[df["CS_RACA"] != 9.0]
df = df[(df['CS_RACA'] != 3.0) & (df['CS_RACA'] != 5.0)]
df["CS_RACA"] = df["CS_RACA"].map(racas)
df = pd.get_dummies(df, columns=['CS_RACA'], prefix='', prefix_sep='')
df
```

CLI_*:
- Considerar 9.0 (ignorado), nan (não preenchido) e 2.0 (negativo) como 0.0
- Manter linhas que possuem ao menos 4 sintomas

```python
df.iloc[:, 18:33] = df.iloc[:, 18:33].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 18:33].sum(axis=1) < 4
df = df[~remover]
df
```

ANT_*:
- Considerar 9.0 (ignorado) e nan (não preenchido) e 2.0 (negativo) como 0.0
- Manter linhas que possui ao menos 1 situação de risco

```python
df.iloc[:, 5:18] = df.iloc[:, 5:18].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 5:18].sum(axis=1) < 1
df = df[~remover]
df
```

CS_ESCOL_N:
- Remover dados ausentes
- Deslocar o valor de escolaridade 1 pra cima e colocar o ignorado no 0.0


```python
df = df[df["CS_ESCOL_N"].notna()]
df.loc[:, "CS_ESCOL_N"] = df["CS_ESCOL_N"] + 1
df.loc[df['CS_ESCOL_N'] == 10.0, 'CS_ESCOL_N'] = 0.0
df.groupby("CS_ESCOL_N").size()
```

Checar mortalidade

```python
#print(sorted(df["NU_IDADE_N"].unique().tolist()))
morte = df[df["EVOLUCAO"].isin([2.0, 3.0])].shape[0]
vivo = df[df["EVOLUCAO"] == 1.0].shape[0]

print(f"total: {morte+vivo} | mortes: {morte} | vivos: {vivo} | taxa: {morte/(morte+vivo)*100:.2f}%")
```

```python
# Mortes por valor de escolaridade
df_temp = df[df['CS_ESCOL_N'].notna()]
resultado = df_temp.groupby('CS_ESCOL_N')['EVOLUCAO'].value_counts().unstack().fillna(0)

for cs_escol_n, row in resultado.iterrows():
  print(f"escolaridade: {cs_escol_n} | mortes: {row[2.0]+row[3.0]} | taxa: {(row[2.0]+row[3.0])/(row[1.0]+row[2.0]+row[3.0])*100:.2f}%")
```
