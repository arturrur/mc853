Ler arquivo csv e manter apenas as colunas necess√°rias

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/arturrur/mc853/refs/heads/main/source/csv/LEPTBR23.csv", usecols=["SEM_NOT", "SG_UF_NOT", "DT_NOTIFIC" , "DT_SIN_PRI", "NU_IDADE_N",
                                                                                                                      "CS_SEXO", "CS_ESCOL_N", "CS_GESTANT", "NDUPLIC_N", "ANT_CB_LAM", "ANT_CB_CRI", "CS_RACA",
                                                                                                                      "ANT_CB_CAI", "ANT_CB_FOS", "ANT_CB_SIN", "ANT_CB_PLA", "ANT_CB_COR",
                                                                                                                      "ANT_CB_ROE", "ANT_CB_GRA", "ANT_CB_TER", "ANT_CB_LIX", 'CLI_FEBRE',
                                                                                                                      'CLI_MIALGI', 'CLI_CEFALE', 'CLI_PROST', 'CLI_CONGES', 'CLI_PANTUR',
                                                                                                                      'CLI_VOMITO', 'CLI_DIARRE', 'CLI_ICTERI','CLI_RENAL', 'CLI_RESPIR',
                                                                                                                      'CLI_CARDIA', 'CLI_HEMOPU', 'CLI_HEMORR', 'CLI_MENING', "ANT_HUMANO",
                                                                                                                      "ANT_ANIMAI", "CLI_DT_ATE", "EVOLUCAO"])
df["DT_NOTIFIC"] = pd.to_datetime(df["DT_NOTIFIC"])
df["DT_SIN_PRI"] = pd.to_datetime(df["DT_SIN_PRI"])
df["CLI_DT_ATE"] = pd.to_datetime(df["CLI_DT_ATE"])
df
```

EVOLUCAO:
- Remover linhas do df que n√£o tenham valor
- Remover linhas que tenham valor ignorado

```python
df = df.dropna(subset=["EVOLUCAO"])
df = df[df["EVOLUCAO"] != 9.0]
df
```

NDUPLIC_N:
- Remover duplicatas
- Remover a coluna NDUPLIC_N, que n√£o tem mais uso

```python
df = df[df["NDUPLIC_N"] != 2.0]
df = df.drop(columns=["NDUPLIC_N"])
df
```

SG_UF_NOT:
- Criar bitmap a partir da coluna SG_UF_NOT
- Deletar coluna SG_UF_NOT

```python
regions ={
    1: "Norte",
    2: "Nordeste",
    3: "Sudeste",
    4: "Sul",
    5: "Centro-Oeste"
}
df["int"] = df["SG_UF_NOT"] // 10
df["region"] = df["int"].map(regions)
df = pd.get_dummies(df, columns=['region'], prefix='', prefix_sep='')
df = df.drop(columns=["int", "SG_UF_NOT"])

print(df)
```

SEM_NOT:
- Reformatar SEM_NOT, que est√° na forma aaaass, para apenas ss (semana).
- Transformar a primeira semana do ano 2024 para a semana 53 do ano 2023.

Obs: Teve uma amostra que tinha o valor de SEM_NOT como 2350, muito provavelmente um erro de digita√ß√£o para 202350, ajustamos seu valor

```python
df.loc[df['SEM_NOT'] == 202401, 'SEM_NOT'] = 202353
df.loc[df['SEM_NOT'] == 2350, 'SEM_NOT'] = 202350
df['SEM_NOT'] = df['SEM_NOT'] % 100
df
```

CLI_DT_ATE e DT_SIN_PRI:
- Criar uma coluna que representa a quantidade de dias entre primeiros sintomas e data de atendimento.
- Tirar dos dados aqueles que t√™m mais de 20 dias entre as duas datas.
- Remover as colunas CLI_DT_ATE e DT_SIN_PRI, que n√£o t√™m mais uso.

```python
df['DIF_SINTOMAS_ATEND'] = (df['CLI_DT_ATE'] - df['DT_SIN_PRI']).dt.days
df = df[(df['DIF_SINTOMAS_ATEND'] <= 20) & (df['DIF_SINTOMAS_ATEND'] >= 0)]
df = df.drop(columns=["CLI_DT_ATE", "DT_SIN_PRI"])

df
```

CS_GESTANT, transformar a coluna em bool
- True se gestante
- False caso contr√°rio

```python
df['CS_GESTANT'] = df['CS_GESTANT'].apply(lambda x: True if x in [1, 2, 3, 4] else False)
df
```

NU_IDADE_N:
- Remover todos que t√™m menos que 1 ano de idade
- Reformatar a coluna de idade para mostrar apenas quantos anos eles t√™m

```python
df = df[df['NU_IDADE_N'] // 1000 == 4]
df.loc[:, 'NU_IDADE_N'] = df['NU_IDADE_N'] % 1000
df
```

CS_RACA:
- Remover dados faltantes
- Remover ra√ßas com poucos dados (amarela e ind√≠gena)
- Criar um bitmap pras ra√ßas

```python
racas = {
    1: "Branca",
    2: "Preta",
    3: "Amarela",
    4: "Parda",
    5: "Ind√≠gena",
    9: "Ignorado"
}

df = df.dropna(subset=["CS_RACA"])
df = df[df["CS_RACA"] != 9.0]
df = df[(df['CS_RACA'] != 3.0) & (df['CS_RACA'] != 5.0)] # a professora fez isso, n√£o sei se devemos fazer tbm ou n√£o
df["CS_RACA"] = df["CS_RACA"].map(racas)
df = pd.get_dummies(df, columns=['CS_RACA'], prefix='', prefix_sep='')
df
```

CLI_*:
- Considerar 9.0 (ignorado) e nan (n√£o preenchido) e 2.0 (negativo) como 0.0
- Manter linhas que possuem ao menos 4 sintomas

```python
df.iloc[:, 19:34] = df.iloc[:, 19:34].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 19:34].sum(axis=1) < 4
df = df[~remover]
df
```

ANT_*:
- Considerar 9.0 (ignorado) e nan (n√£o preenchido) e 2.0 (negativo) como 0.0
- Manter linhas que possui ao menos 1 situa√ß√£o de risco

```python
df.iloc[:, 6:20] = df.iloc[:, 6:20].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 6:20].sum(axis=1) < 1
df = df[~remover]
df
```

CS_ESCOL_N:
- Remover dados ausentes, ignorados e "n√£o se aplica"
- Ao remover o "n√£o se aplica" estamos retirando dos dados todas as pessoas com menos de 7 anos

```python
df = df[df["CS_ESCOL_N"].notna()]
df = df[~df["CS_ESCOL_N"].isin([9.0, 10.0])]

```

```python
#print(sorted(df["NU_IDADE_N"].unique().tolist()))
morte = df[df["EVOLUCAO"].isin([2.0, 3.0])].shape[0]
vivo = df[df["EVOLUCAO"] == 1.0].shape[0]

print(f"total: {morte+vivo} | mortes: {morte} | vivos: {vivo} | taxa: {morte/(morte+vivo)*100:.2f}%")
```

```python
# Mortes por valor de escolaridade
df_temp = df[df['CS_ESCOL_N'].notna()]
resultado = df_temp.groupby('CS_ESCOL_N')['EVOLUCAO'].value_counts().unstack().fillna(0)

for cs_escol_n, row in resultado.iterrows():
  print(f"escolaridade: {cs_escol_n} | mortes: {row[2.0]+row[3.0]} | taxa: {(row[2.0]+row[3.0])/(row[1.0]+row[2.0]+row[3.0])*100:.2f}%")
```
