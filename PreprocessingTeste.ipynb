Fazer literalmente a mesma limpeza que fizemos para os dados de treinamento

```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/arturrur/mc853/refs/heads/using-KNN-imputer/data/raw/LEPTBR19.csv", usecols=["SEM_NOT", "SG_UF_NOT", "DT_SIN_PRI", "NU_IDADE_N", "CS_RACA",
                                                                                                                      "CS_SEXO", "CS_ESCOL_N", "CS_GESTANT", "NDUPLIC_N", "ANT_CB_LAM", "ANT_CB_CRI",
                                                                                                                      "ANT_CB_CAI", "ANT_CB_FOS", "ANT_CB_SIN", "ANT_CB_PLA", "ANT_CB_COR",
                                                                                                                      "ANT_CB_ROE", "ANT_CB_GRA", "ANT_CB_TER", "ANT_CB_LIX", "ANT_HUMANO", "ANT_ANIMAI",
                                                                                                                      'CLI_FEBRE', 'CLI_MIALGI', 'CLI_CEFALE', 'CLI_PROST', 'CLI_CONGES', 'CLI_PANTUR',
                                                                                                                      'CLI_VOMITO', 'CLI_DIARRE', 'CLI_ICTERI','CLI_RENAL', 'CLI_RESPIR',
                                                                                                                      'CLI_CARDIA', 'CLI_HEMOPU', 'CLI_HEMORR', 'CLI_MENING',
                                                                                                                      "CLI_DT_ATE", "EVOLUCAO"])

df["DT_SIN_PRI"] = pd.to_datetime(df["DT_SIN_PRI"])
df["CLI_DT_ATE"] = pd.to_datetime(df["CLI_DT_ATE"])
df
```

```python
df = df.dropna(subset=["EVOLUCAO"])
df = df[df["EVOLUCAO"] != 9.0]
df.loc[:, "EVOLUCAO"] = df["EVOLUCAO"] - 1
df.loc[df['EVOLUCAO'] == 2.0, 'EVOLUCAO'] = 1.0
df = df.rename(columns={'EVOLUCAO': "Desfecho"})
df
```

```python
df = df[df["NDUPLIC_N"] != 2.0]
df = df.drop(columns=["NDUPLIC_N"])
df
```

```python
regions ={
    1: "Norte",
    2: "Nordeste",
    3: "Sudeste",
    4: "Sul",
    5: "Centro-Oeste"
}
df["int"] = df["SG_UF_NOT"] // 10
df["region"] = df["int"].map(regions)
df = pd.get_dummies(df, columns=['region'], prefix='', prefix_sep='')
df = df.drop(columns=["int", "SG_UF_NOT"])

df[['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']] = df[['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']].astype(int)

df
```

```python
df["SEM_NOT"] = df["SEM_NOT"].apply(lambda x: (x % 100))
df = df.rename(columns={'SEM_NOT': 'Semana_epi'})
df
```

```python
df['Intervalo'] = (df['CLI_DT_ATE'] - df['DT_SIN_PRI']).dt.days
df = df[(df['Intervalo'] <= 20) & (df['Intervalo'] >= 0)]
df = df.drop(columns=["CLI_DT_ATE", "DT_SIN_PRI"])
df
```

```python
df = df[df["CS_SEXO"].isin(['M', 'F'])]
df.loc[:,'CS_SEXO'] = df['CS_SEXO'].map({'M': 0, 'F': 1})
df = df.rename(columns={'CS_SEXO': "Sexo"})
df
```

```python
df.loc[:,'CS_GESTANT'] = df['CS_GESTANT'].apply(lambda x: 1 if x in [1, 2, 3, 4] else 0)
df = df.rename(columns={'CS_GESTANT': 'Gestante'})
df
```

```python
df = df[df['NU_IDADE_N'] // 1000 == 4]
df.loc[:, 'NU_IDADE_N'] = df['NU_IDADE_N'] % 1000
df = df[df['NU_IDADE_N'] >= 7]
df = df.rename(columns={'NU_IDADE_N': 'Idade'})
df
```

```python
racas = {
    1: "Branca",
    2: "Preta",
    3: "Amarela",
    4: "Parda",
    5: "Indigena",
    9: "Ignorado"
}

df = df.dropna(subset=["CS_RACA"])
df = df[df["CS_RACA"] != 9.0]
df = df[(df['CS_RACA'] != 3.0) & (df['CS_RACA'] != 5.0)]
df["CS_RACA"] = df["CS_RACA"].map(racas)
df = pd.get_dummies(df, columns=['CS_RACA'], prefix='', prefix_sep='')
df[['Preta', 'Branca', 'Parda']] = df[['Preta', 'Branca', 'Parda']].astype(int)
df
```

```python
df.iloc[:, 18:33] = df.iloc[:, 18:33].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 18:33].sum(axis=1) < 4
df = df[~remover]
df = df.rename(columns={
    'CLI_FEBRE': 'Febre',
    'CLI_MIALGI': 'Mialgia',
    'CLI_CEFALE': 'Cefaleia',
    'CLI_PROST': 'Prostracao',
    'CLI_CONGES': 'Congestao_conjuntival',
    'CLI_PANTUR': 'Dor_panturrilha',
    'CLI_VOMITO': 'Vomito',
    'CLI_DIARRE': 'Diarreia',
    'CLI_ICTERI': 'Ictericia',
    'CLI_RENAL': 'Insuficiencia_renal',
    'CLI_RESPIR': 'Alteracoes_respiratorias',
    'CLI_CARDIA': 'Alteracoes_cardiacas',
    'CLI_HEMOPU': 'Hemorragia_pulmonar',
    'CLI_HEMORR': 'Outras_hemorragias',
    'CLI_MENING': 'Meningismo'
})
df
```

```python
df.iloc[:, 5:18] = df.iloc[:, 5:18].replace([9.0, 2.0], 0.0).fillna(0.0)
remover = df.iloc[:, 5:18].sum(axis=1) < 1
df = df[~remover]
df = df.rename(columns={
    'ANT_CB_LAM': 'Lama_enchente',
    'ANT_CB_CRI': 'Criacao_animal',
    'ANT_CB_CAI': 'Caixa_dagua',
    'ANT_CB_FOS': 'Fossa_esgoto',
    'ANT_CB_SIN': 'Sinal_roedor',
    'ANT_CB_PLA': 'Plantio_colheita',
    'ANT_CB_COR': 'Corpos_dagua',
    'ANT_CB_ROE': 'Roedores_diretamente',
    'ANT_CB_GRA': 'Armazenamento_graos',
    'ANT_CB_TER': 'Terreno_baldio',
    'ANT_CB_LIX': 'Lixo_entulho',
    'ANT_HUMANO': 'Caso_anterior_humano',
    'ANT_ANIMAI': 'Caso_anterior_animal'
})
df
```

```python
df.loc[:, "CS_ESCOL_N"] = df["CS_ESCOL_N"].mask(df["CS_ESCOL_N"] == 9.0, np.nan)
df = df.rename(columns={'CS_ESCOL_N': 'Escolaridade'})
df
```

```python
# Mover Desfecho para última coluna
colunas = [col for col in df.columns if col != "Desfecho"] + ["Desfecho"]
df = df[colunas]
# Mudar de float para inteiro mantendo NaN como NaN
df = df.astype('Int64')
df
```

```python
morte = df[df["Desfecho"] == 1.0].shape[0]
vivo = df[df["Desfecho"] == 0.0].shape[0]

print(f"total: {morte+vivo} | mortes: {morte} | vivos: {vivo} | taxa: {morte/(morte+vivo)*100:.2f}%")
```

```python
# Mortes por valor de escolaridade
df_temp = df[df['Escolaridade'].notna()]
resultado = df_temp.groupby('Escolaridade')['Desfecho'].value_counts().unstack().fillna(0)

for escolaridade, row in resultado.iterrows():
  print(f"escolaridade: {escolaridade} | mortes: {row[1.0]} | taxa: {(row[1.0])/(row[0.0])*100:.2f}%")
```
